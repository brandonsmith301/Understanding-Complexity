# <p id="introduction" style="font-family:JetBrains Mono; font-weight:normal; letter-spacing: 2px; color:#000000; font-size:140%; text-align:left;padding: 0px; border-bottom: 3px solid #F1F6F9">1. Introduction</p>

In this article, we will explore the concept of time complexity, its various types, and how to gain a better understanding of it. Time complexity plays a crucial role in evaluating the efficiency of algorithms and helps us make informed decisions when choosing the best approach to solve a particular problem. Michael Sambol's quick introduction to time complexity and its significance is well-explained in his video [Big-O notation in 5 minutes](https://www.youtube.com/watch?v=__vX2sjlpXU&ab_channel=MichaelSambol). 

When creating and implementing a program, the primary focus should be on delivering accurate and trustworthy outcomes. We want our medical diagnoses to be precise, the climate control systems in our homes to maintain the correct temperature, and to prevent data breaches in secure systems.

In certain cases, performance is a vital component of a program's reliability. This is particularly true for real-time programs, such as traffic management systems that must adjust traffic signals in response to real-time conditions. Performance also influences the practicality of numerous non-real-time programs. For instance, the speed at which a search engine returns results is an essential factor when evaluating its effectiveness. Users care about the time it takes to load a website on their devices, and astronomers are concerned with how long it takes to process massive amounts of data from telescopes.

Creating efficient programs can be challenging. The simplest solution is frequently not the most effective one. Highly efficient algorithms tend to use clever techniques that can make them harder to comprehend. As a result, developers often increase a program's **conceptual complexity** to decrease its **computational complexity** <a href="#1">[1]</a>. To do this sensibly, it's crucial to learn how to estimate a program's computational complexity accurately. This subject is the primary focus of this artical.

# <p id="typesofcomplexity" style="font-family:JetBrains Mono; font-weight:normal; letter-spacing: 2px; color:#000000; font-size:140%; text-align:left;padding: 0px; border-bottom: 3px solid #F1F6F9">2. Types of Complexity</p>

When selecting an algorithm, it's essential to consider three distinct scenarios of time complexity <a href="#11">[11]</a>:

- **Best case** - This represents the optimal running time an algorithm can achieve when the inputs are as favorable as possible <a href="#1">[1]</a>. In this situation, the algorithm performs at its highest efficiency.


- **Average case** - The average-case complexity of an algorithm refers to the time utilised by the algorithm, averaged across all potential inputs <a href="#2">[2]</a>. This metric provides a more comprehensive view of the algorithm's performance in typical situations.


- **Worst case** - This scenario measures the resources (e.g., running time, memory) an algorithm requires for an input of arbitrary size. The worst-case complexity highlights the upper bound on an algorithm's performance, helping to identify potential bottlenecks or inefficiencies when dealing with large inputs <a href="#3">[3]</a>. 


Asymptotic analysis determines the best, average, and worst-case scenarios for an algorithm. It is done by utilising a mathematical technique which assesses the efficiency of algorithms in terms of their run-time. Asymptotic analysis concentrates on the input size while keeping all other factors constant.

In practical terms, asymptotic analysis calculates the running time of a process using mathematical units of computation. This means that one process might have a running time of $f(n)$, while another process could have a running time of $g(n^2)$. When $n$ is relatively small, both processes will exhibit similar running times. However, as $n$ grows, the running time of the first process will increase linearly, while the running time of the second process will rise exponentially <a href="#4">[4]</a>. We won't continue into asymptotic analysis as it is out of scope for what I am trying to accomplish here. However, for further reading, you can refer to <a href="#12">[12]</a>.

# <p id="ConstantComplexity" style="font-family:JetBrains Mono; font-weight:normal; letter-spacing: 2px; color:#000000; font-size:140%; text-align:left;padding: 0px; border-bottom: 3px solid #F1F6F9">2.1 Constant Complexity</p>

Denoted by $O(1)$, which means the algorithm takes the same amount of time to process the input $n$, no matter how large or small it is. A limited number of intriguing programs fall into the constant complexity class, but every program contains components that fit this category. For instance as presented by <a href="#1">[1]</a>, determining the length of a Python list or multiplying two floating-point numbers are examples of operations with constant complexity.

The example below represents a Python function which runs at constant time, regardless of the size of the input list, the function will always take the same amount of time to return the first element. It doesn't matter if lst contains 10 elements or 100,000 elements; the running time remains constant.

In simple terms, to determine if an algorithms running time is constant, you can simpily look for operations that don't depend on the input size, such as accessing an array value such as <code>get_first_value</code> or performing basic arithmetic.

```python
def get_first_value(lst):
    return lst[0]
```

```python
n = list(range(1, 100001)) 
running_times = [1] * len(n) 

plt.plot(n, running_times)
plt.xlabel("Input Size $(n)$")
plt.ylabel("Running Time")
plt.title("Constant Time Complexity - $O(1)$")
plt.show()
```

# <p id="LogarithmicComplexity" style="font-family:JetBrains Mono; font-weight:normal; letter-spacing: 2px; color:#000000; font-size:140%; text-align:left;padding: 0px; border-bottom: 3px solid #F1F6F9">2.2 Logarithmic Complexity</p>

Indicated by $O(log\ n)$, this notation implies that an algorithm's running time grows logarithmically with the input size $(n)$. In other words, as the input size $(n)$ increases, the algorithm's running time does not grow linearly but rather at a slower, logarithmic rate.

In <a href="#5">[5]</a>, the performance of Binary Search is analysed, revealing it as an efficient search algorithm with logarithmic performance in both average and worst-case scenarios.

If the running time increases slowly as the input size grows, it's likely to be $O(log\ n)$. In the context of algorithms, a logarithmic algorithm often involves repeatedly dividing the problem size or input by a constant factor until the problem is solved, like Binary Search.

```python
def binary_search(lst, target):
    low, high = 0, len(lst) - 1

    while low <= high:
        mid = (low + high) // 2
        if lst[mid] == target:
            return mid
        elif lst[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1
```

```python
input_sizes = np.linspace(1, 100001)
running_times = np.log2(input_sizes)

plt.plot(input_sizes, running_times)
plt.xlabel("Input Size $(n)$")
plt.ylabel("Running Time")
plt.title("Logarithmic Time Complexity - $O(log\ n)$")
plt.show()
```

### Further examples on Logarithmic Complexity:

We will consider another two different functions which have an logarithmic complexity and analyse them both. 

Let us start off with <code>count_dights</code>:

```python
def count_digits(number):
    """Assumes number is a nonnegative int
    Returns the count of digits in the number"""
    count = 0
    while number > 0:
        number = number // 10
        count += 1
    return count
```

Like the example in <a href="#1">[1]</a>, we can examine the while loop in the code to understand the time complexity, since there are no function or method calls to consider.

<code>count_digits</code> uses a loop that repeatedly divides the number by 10 until it reaches 0. This means that the number of iterations of the loop is equal to the number of digits in the input number, which is directly related to the logarithm of the number.

Now consider another function <code>max_depth</code>:

```python
def max_depth(node_count):
    """Assumes node_count is a nonnegative int
    Returns the maximum depth of a balanced binary search tree with node_count nodes"""
    depth = 0
    while node_count > 0:
        node_count = node_count // 2
        depth += 1
    return depth
```

The loop in <code>max_depth</code> continues to iterate until the total variable (node_count) is reduced to 0, dividing by 2 at each step. 

Similar to our <code>count_digits</code> example, the number of iterations depends on the depth of the tree, which is logarithmic in the number of nodes (base 2) where as the previous example it was (base 10).

What is interesting about these two examples is the similarities they share, which is that they both involve division, leading to a logarithmic relationship between the input size and the number of iterations. This is a key indicator that the time complexity of an algorithm might be $O(log(n))$. So, when you see a division operation in a loop that reduces the input size, it's often a sign that the time complexity is logarithmic <a href="#6">[6]</a>. 

# <p id="LinearComplexity" style="font-family:JetBrains Mono; font-weight:normal; letter-spacing: 2px; color:#000000; font-size:140%; text-align:left;padding: 0px; border-bottom: 3px solid #F1F6F9">2.3 Linear Complexity</p>

Linear complexity, represented by $O(n)$, implies that an algorithm processes each element in a given input, such as a list or a sequence, a constant number of times <a href="#1">[1]</a>. The running time of the algorithm increases proportionally with the growth of the input size ($n$).

An example of an algorithm which runs $O(n)$ is Linear Search. The reason is it iterates through each value in a list, comparing the current value with the target value. Once the match is found , the algorithm will terminate and return the index of the matching value.

```python
def linear_search(lst, target):
    for i, value in enumerate(lst):
        if value == target:
            return i
    return -1
```

```python
input_sizes = np.linspace(1, 100001)
running_times = input_sizes

plt.plot(input_sizes, running_times)
plt.xlabel("Input Size $(n)$")
plt.ylabel("Running Time")
plt.title("Linear Time Complexity - $O(n)$")
plt.show()
```

### Further examples on Linear Complexity:

Let us consider further examples on functions which have linear complexity. We will start off with summing the values of a list.

```python
def sum_values(lst):
    """Assumes lst is a list of integers.
    Returns the sum of the values in lst."""
    total = 0
    for number in lst:
        total += number
    return total
```

<code>sum_values</code> is linear in the length of the lst, i.e. $O(len(lst))$. The reason for its linear complexity is that it iterates through each element in the list exactly once, performing a constant number of operations (addition) on each value. 

Now is a good time to reflect back on the function <code>get_first_value</code> which had constant time complexity. Performing basic arithmetic is constant, however, in our example above, it is no longer constant because the time complexity now depends on the length of the input list or more formally speaking the size of $n$. 

Let us consider another example, which is provided in <a href="#1">[1]</a>. The reason I chose exact example is because often it can be assumed that in order to have linear complexity, one must have a loop but this isn't the case.

```python
def factorial(num):
    """Assumes num is a positive integer
     Returns the factorial of num, denoted as num!"""
    if num == 1:
        return 1
    else:
        return num * factorial(num - 1)
```

The `factorial` function does not have any loops, but it uses recursion to compute the factorial. 

Recursion works in a similar way to loops, except that the future iterations depend on the previous iteration. For example, `factorial(num)`, `factorial(num - 1)`, `factorial(num - 2),` ..., `factorial(1)`.

Therefore, just like <code>sum_values</code>, the time complexity is $O(num)$. This is because the length of the sequence, which refers to the number of times `factorial` is called is equal to the input size. Which means that the time complexity will grow linearly as the size of (num) increases.

# <p id="Log-LinearComplexity" style="font-family:JetBrains Mono; font-weight:normal; letter-spacing: 2px; color:#000000; font-size:140%; text-align:left;padding: 0px; border-bottom: 3px solid #F1F6F9">2.4 Log-Linear Complexity</p>

Log-linear complexity, denoted by $O(n\log n)$, signifies that the running time of an algorithm is affected by both the size of the input $(n)$ and the logarithm of the input size <a href="#7">[7]</a>. In other words, as the input size $(n)$ grows, the algorithm's running time increases at a rate that is proportional to the product of the input size and the logarithm of the input size.

In **Task 6.3D** we are required to find sorting algorithms that meet $O(n\log n)$ complexity on average. There are three different sorting algorithms listed in <a href="#7">[7]</a> that meet his time complexity on average.

We will only discuss one called Merge sort. This sorting algorithm is a divide-and-conquer algorithm that recursively divides the input list into halves, sorts the halves, and then merges the sorted halves back together <a href="#1">[1]</a>. 

The reason for its log-linear complexity is that it needs to perform the merge operation for each level of the recursion tree, and there are $\log_2 n$ levels in the tree. At each level, the merge operation takes linear time, i.e. $O(n)$. Hence, the average-case complexity being $O(n\log n)$.

In the visual example, I have decided to plot the log-linear, linear, and logarithmic time complexities together because log-linear on its own looks very similar to linear. However, when plotted together, the differences between the three become clear, and they are significant.

```python
input_sizes = np.linspace(1, 100001)

running_times_log = input_sizes * np.log(input_sizes)
running_times_linear = input_sizes
running_times_loglog = np.log(input_sizes)

plt.plot(input_sizes, running_times_log, label="Log-linear - $O(n\log n)$")
plt.plot(input_sizes, running_times_linear, label="Linear - $O(n)$")
plt.plot(input_sizes, running_times_loglog, label="Logarithmic - $O(\log n)$")

plt.xlabel("Input Size $(n)$")
plt.ylabel("Running Time")
plt.title("Comparison of Log-linear, Linear and Logarithmic")

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Lets take a look of an implementation of merge sort algorithm (if interested in how it is applied through various languages you can visit <a href="https://www.geeksforgeeks.org/merge-sort/">Merge Sort Algorithm</a>):

```python
def merge_sort(lst):
    if len(lst) <= 1:
        return lst
    
    mid = len(lst) // 2
    left = merge_sort(lst[:mid])
    right = merge_sort(lst[mid:])
    
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    
    return result
```

By combining our understanding of linear and logarithmic time complexity, we can apply similar analysis to log-linear time complexity.

In the implementation from <a href="#8">[8]</a>, `merge_sort` recursively divides the input list into halves (left and right) until the base case is reached (list of length 1 or less). Then, the merge function is used to combine the sorted halves. Since the `merge` operation takes linear time ($O(n)$) and the recursion tree has a depth of $\log_2 n$, the overall time complexity of `merge_sort` is $O(n\log n)$.

# <p id="PolynomialComplexity" style="font-family:JetBrains Mono; font-weight:normal; letter-spacing: 2px; color:#000000; font-size:140%; text-align:left;padding: 0px; border-bottom: 3px solid #F1F6F9">2.5 Polynomial Complexity</p>

Polynomial complexity, represented by $O(n^k)$, implies that the running time of an algorithm increases with the input size $(n)$ raised to the power of a constant $(k)$ <a href="#7">[7]</a>. Quadratic complexity $(k=2)$ is a common example of polynomial complexity.

Consider the following function (this is not efficient at all, there are faster ways of doing this which involves using sets):

```python
def check_value(lst1, lst2):
    """Assuming they are the same length, checks if value is in 
    lst1 and list2 then appends to new list confirmed"""
    confirmed = []
    for i in lst1:
        for k in lst2:
            if i == k:
                confirmed.append(i)
    return confirmed
```

In `check_value` the function has a nested loop, with the outer loop interating $O(len(lst1))$ times and the inner loop iterating $O(len(lst2))$ times. Therefore, if we let $n_1$ be `lst1` and $n_2$ `lst2`, the overall complexity is $(len(n_1) \times len(n_2))$.

To visualise what this would look like, you can view the plot below. 

```python
input_sizes = np.linspace(1, 1001)
running_times = input_sizes**2

plt.plot(input_sizes, running_times)
plt.xlabel("Input Size $(n)$")
plt.ylabel("Running Time")
plt.title("Quadratic Time Complexity - $O(n^2)$")
plt.show()
```

# <p id="ExponentialComplexity" style="font-family:JetBrains Mono; font-weight:normal; letter-spacing: 2px; color:#000000; font-size:140%; text-align:left;padding: 0px; border-bottom: 3px solid #F1F6F9">2.6 Exponential Complexity</p>

This type of complexity is denoted by $O(c^n)$. The running time of the algorithm increases exponentially with the input size $(n)$ and the base $(c)$ of the exponential function <a href="#7">[7]</a>. As discused in <a href="#1">[1]</a> many important problems are exponential meaning solving them can require time that is exponential in the size of the input.

```python
input_sizes = np.arange(1, 11)
running_times = 2**input_sizes - 1

plt.plot(input_sizes, running_times)
plt.xlabel("Input Size $(n)$")
plt.ylabel("Running Time")
plt.title("Exponential Complexity - $O(c^n)$")
plt.show()
```

One such example is the <a href="https://leetcode.com/discuss/general-discussion/470927/tower-of-hanoi-algorithm">Tower of Hanoi</a> by using a recursion. The Tower of Hanoi is a classic puzzle where you have three rods and a number of disks of different sizes <a href="#9">[9]</a>. The objective is to move the entire stack of disks from the first rod to the last, obeying the following rules:
1. Only one disk can be moved at a time.
2. Each move consists of taking the top disk from one of the stacks and placing it on top of another stack or an empty rod.
3. No disk may be placed on top of a smaller disk.



```python
def tower_of_hanoi(n, source, target, auxiliary):
    """Assumes n is a positive integer representing the number of disks,
    source, target, and auxiliary are strings representing rod names.
    Solves the Tower of Hanoi problem and prints the steps."""
    if n > 0:
        # Move n-1 disks from source to auxiliary, using target as auxiliary
        tower_of_hanoi(n - 1, source, auxiliary, target)
        
        # Move the nth disk from source to target
        print(f"Move disk {n} from {source} to {target}")
        
        # Move the n-1 disks from auxiliary to target, using source as auxiliary
        tower_of_hanoi(n - 1, auxiliary, target, source)
```

Let's break down the Tower of Hanoi algorithm and explain why its complexity is $O(2^n)$.

First of all, for each disk, $n$, the algorithm makes two recursive calls $(n-1)$ disks:
1. The first call moves $n-1$ disk from the source rod to the auxiliary rod, using the target rod as the auxiliary.
2. The second call moves $n-1$ disks from the auxiliary rod to the target rod, using the source rod as the auxiliary.

The base case occurs when $n$ equals 0. In this scenario, no further recursion is needed, and the function terminates.

The depth of the recursion for the algorithm is $n$, which means the number of calls doubles for each additional disk.

When we run the algorithm, step by step: 
- For 1 disk $(n=1)$, there is 1 move $(2^1 - 1 = 1)$.

```python
tower_of_hanoi(1, "A", "C", "B")
```

- For 2 disks $(n=2)$, there are 3 moves $(2^2 - 1 = 3)$.

```python
tower_of_hanoi(2, "A", "C", "B")
```

- For 3 disks $(n=3)$, there are 7 moves $(2^3 - 1 = 7)$.

```python
tower_of_hanoi(3, "A", "C", "B")
```

As we continue with more disks, the pattern holds:
- For 4 disks $(n=4)$, there are 15 moves $(2^4 - 1 = 15)$.
- For 5 disks $(n=5)$, there are 31 moves $(2^5 - 1 = 31)$.

As you can see, the number of moves required to solve the problem consistently follows the pattern of $2^n - 1$. In the exponential complexity notation $O(c^n)$, the base $c$ is 2. The recursion reduces the problem size by 1 at each level, so $n$ becomes $n-1$ in the recursive calls. This pattern confirms that the complexity of the algorithm is $O(2^n)$.
